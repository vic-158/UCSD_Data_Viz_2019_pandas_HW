## Gender Demographics

```python
#Extract the value_counts for purchase_data['Gender'] as its own series and create new DF
gender_series = uniquePlayersDF['Gender'].value_counts()
gender_demo = pd.DataFrame(gender_series)

#Calculate list of percentages of each gender and add to gender DF as new column
gender_pct = []
for num in gender_demo['Gender']:
    gender_pct.append('{:,.2f}'.format((num/gender_demo['Gender'].sum())*100))
gender_demo['Percentage of Players'] = gender_pct
gender_demo
```

## Age Demographics

```python
bins  = [0,10,15,20,25,30,35,40,100]
group_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

purchase_data['Age Bins'] = pd.cut(purchase_data['Age'], bins, labels=group_names,right = False)
purchase_data.head()
```

```python
age_demographics = purchase_data.groupby('Age Bins').count()
age_summary = age_demographics[['Purchase ID']]
```

```python
age_pct = []
for items in age_summary['Purchase ID']:
    age_pct.append('{:,.2f}'.format(items/age_summary['Purchase ID'].sum()*100))
age_summary['Percentage of Players'] = age_pct
age_summary
```
